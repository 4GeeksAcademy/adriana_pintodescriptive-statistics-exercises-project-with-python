# 2. Media y Desv. Estandar en algunas Distribuciones de Probabilidad.

```python
# Parámetros de la distribución normal
mu = 50
sigma = 10
size = 1000  # Tamaño de la muestra

# Generar datos
data_normal = np.random.normal(mu, sigma, size)

# Cálculo de media y desviación estándar
mean_normal = np.mean(data_normal)
std_normal = np.std(data_normal, ddof=1)  # ddof=1 para la muestra

mean_normal, std_normal
```

```python
# Parámetros de la distribución binomial
n = 20  # Número de ensayos
p = 0.4  # Probabilidad de éxito
size = 1000

# Generar datos
data_binomial = np.random.binomial(n, p, size)

# Cálculo de media y desviación estándar
mean_binomial = np.mean(data_binomial)
std_binomial = np.std(data_binomial, ddof=1)

mean_binomial, std_binomial
```

```python
# Parámetros de la distribución Chi-Cuadrado
k = 5  # Grados de libertad
size = 1000

# Generar datos
data_chi2 = np.random.chisquare(k, size)

# Cálculo de media y desviación estándar
mean_chi2 = np.mean(data_chi2)
std_chi2 = np.std(data_chi2, ddof=1)

mean_chi2, std_chi2
```

```python
# Parámetros de la distribución uniforme
a = 0   # Límite inferior
b = 10  # Límite superior
size = 1000  # Número de muestras

# Generar datos
data_uniform = np.random.uniform(a, b, size)

# Cálculo de media y desviación estándar
mean_uniform = np.mean(data_uniform)
std_uniform = np.std(data_uniform, ddof=1)

mean_uniform, std_uniform
```

# 3. Graficos de Media y Desv. Estandar en algunas Distribuciones de Probabilidad.

```python
#loc 
#scale

# Generar datos de una normal con media=0 y sigma=1
data = np.random.normal(loc=0, scale=1, size=1000)

# Graficar la distribución
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black')

# Superponer la PDF de la normal teórica
x = np.linspace(-4, 4, 1000)
pdf = stats.norm.pdf(x, loc=0, scale=1)
plt.plot(x, pdf, 'r', linewidth=2)

# Superponer la PDF de la normal teórica
x = np.linspace(-4, 4, 1000)
pdf = stats.norm.pdf(x, loc=0, scale=1)
plt.plot(x, pdf, 'r', linewidth=2)
```

```python
#df

# Generar datos de una distribución Chi-Cuadrado con 3 grados de libertad
chi_square_data = np.random.chisquare(df=3, size=1000)

# Graficar la distribución
plt.hist(chi_square_data, bins=30, density=True, alpha=0.6, color='green', edgecolor='black')

# Superponer la PDF teórica
x = np.linspace(0, 15, 1000)
pdf = stats.chi2.pdf(x, df=3)
plt.plot(x, pdf, 'r', linewidth=2)

plt.title("Distribución Chi-Cuadrado (df=3)")
plt.xlabel("Valores")
plt.ylabel("Densidad de probabilidad")
plt.show()
```

```python
#a
#b

# Generar datos de una distribución uniforme
data_uniform = np.random.uniform(a, b, size)

# Graficar la distribución
plt.hist(data_uniform, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black')

# Superponer la PDF teórica
x = np.linspace(a, b, 1000)
pdf = np.ones_like(x) / (b - a)  # Función de densidad de probabilidad de la distribución uniforme
plt.plot(x, pdf, 'r', linewidth=2)

plt.title("Distribución Uniforme (a=0, b=10)")
plt.xlabel("Valores")
plt.ylabel("Densidad de probabilidad")
plt.show()
```
