```python
import statistics as stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

# Estadística Descriptiva en Python

La estadística descriptiva es una rama de la estadística que se encarga de recolectar, analizar, interpretar y presentar datos de manera organizada y efectiva. Su principal objetivo es proporcionar resúmenes simples y comprensibles acerca de las características principales de un conjunto de datos, sin llegar a hacer inferencias o predicciones sobre una población más amplia.

```python
# Datos de ejemplo
data = [10, 20, -15, 0, 50, 10, 5, 100]
```

```python
len(data)
```

# Medidas de Tendencia Central

Las medidas de tendencia central describen cómo se centralizan o agrupan los datos en un conjunto. Son esenciales en estadística y análisis de datos porque nos proporcionan un resumen de la información.

```python
# Media
mean = stats.mean(data)
print(f"Media: {mean}")

# Mediana
median = stats.median(data)
print(f"Mediana: {median}")

# Moda
mode = stats.mode(data)
print(f"Moda: {mode}")
```

```python
# Visualización
plt.figure(figsize=(8,5))
sns.histplot(data, bins=4, kde=True, color='blue')
plt.axvline(mean, color='red', linestyle='dashed', label=f"Media: {mean}")
plt.axvline(median, color='green', linestyle='dashed', label=f"Mediana: {median}")
plt.axvline(mode, color='purple', linestyle='dashed', label=f"Moda: {mode}")
plt.legend()
plt.title("Distribución de los datos y medidas de tendencia central")
plt.show()
```

# Medidas de Dispersión

Las medidas de dispersión describen cómo de variados son los datos en un conjunto. Nos muestran cuánto se extienden o varían los datos alrededor de la media.

```python
# Rango
range_ = max(data) - min(data)
print(f"Rango: {range_}")
```

```python
# Varianza y desviación estándar
variance = stats.variance(data)
std = stats.stdev(data)
print(f"Varianza: {variance}")
print(f"Desviación estándar: {std}")
```

```python
# Visualización
plt.figure(figsize=(8,5))
sns.boxplot(x=data, color="orange")
plt.title("Diagrama de Caja - Medidas de Dispersión")
plt.show()
```

# Medidas de Posición

Las medidas de posición nos indican la ubicación de un valor dentro de un conjunto de datos.

Percentiles y cuantiles (percentiles and quantiles)

Son medidas que tratan sobre cómo se puede dividir un conjunto de datos en partes específicas. Estas medidas se utilizan para entender y describir la distribución de datos.

1. Percentil: Divide un conjunto de datos en 100 partes iguales. El k-ésimo percentil indica el valor bajo el cual cae el k% de las observaciones.
2. Cuantil: Divide un conjunto de datos en partes iguales, dependiendo del tipo. Los cuartiles dividen los datos en cuatro partes, los quintiles en cinco, etcétera.

```python
# Creamos un conjunto de datos de ejemplo (notas de estudiantes)
datos = np.array([55, 60, 63, 67, 70, 73, 75, 78, 80, 85, 87, 90, 92, 95, 98])
datos
```

```python
len(datos)
```

```python
type(datos)
```

```python
# Cálculo de percentiles
percentil_25 = np.percentile(datos, 25)  # Primer cuartil (Q1)
percentil_50 = np.percentile(datos, 50)  # Mediana o segundo cuartil (Q2)
percentil_75 = np.percentile(datos, 75)  # Tercer cuartil (Q3)
```

```python
percentil_10 = np.percentile(datos, 10)  # Primer cuartil (Q1)
```

```python
percentil_10
```

```python
print(f"Percentil 25 (Q1): {percentil_25}")
print(f"Percentil 50 (Mediana, Q2): {percentil_50}")
print(f"Percentil 75 (Q3): {percentil_75}")
```

```python
# Cálculo de cuantiles
cuartiles = np.quantile(datos, [0.25, 0.5, 0.75])  # Cuartiles (Q1, Q2, Q3)
```

```python
print(f"Cuartiles: {cuartiles}")
```

```python
# Cálculo con Pandas
df = pd.DataFrame(datos, columns=["Notas"])
df_cuartiles = df.quantile([0.25, 0.5, 0.75])  # Cuartiles usando Pandas
```

# Medidas de Forma

Las medidas de forma describen cómo se distribuyen los valores en un conjunto de datos en relación con la tendencia central.

![Texto alternativo](skewness.png)

```python
#Si la asimetría es 0 (o cercana a 0): La distribución es simétrica, como en una distribución normal.
#Si la asimetría es positiva (>0): La distribución tiene una cola larga a la derecha (asimetría positiva). Esto significa que hay valores atípicos más grandes que elevan la media.
#Si la asimetría es negativa (<0): La distribución tiene una cola larga a la izquierda (asimetría negativa). Indica que hay valores bajos que empujan la media hacia la izquierda.

# Asimetría (Skewness)
skewness = skew(data)
print(f"Asimetría (Skewness): {skewness}")
```

![Texto alternativo](tipos-curtosis.jpg)

```python
#Curtosis normal (Mesocúrtica) → k ≈ 0
#Curtosis alta (Leptocúrtica) → k > 0
#Curtosis baja (Platicúrtica) → k < 0

# Curtosis (Kurtosis)
kurt = kurtosis(data)
print(f"Curtosis: {kurt}")
```

```python
# Visualización
plt.figure(figsize=(8,5))
sns.histplot(data, bins=5, kde=True, color='blue')
plt.title("Distribución de datos - Medidas de Forma")
plt.show()
```
